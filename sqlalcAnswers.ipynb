```python
import sqlalchemy as sql
import pandas as pd
```

```python
mysql_url = 'mysql+mysqlconnector://root:anayak_98@localhost:3306/twitterdb'
engine = sql.create_engine(mysql_url)
connection = engine.connect()
metadata = sql.MetaData()
metadata
```

```python
tweets = sql.Table('tweets', metadata, autoload = True, autoload_with = engine)
tweets
```

```python
# Print the column names
col = tweets.columns
print(col.keys())
```

```python
# Equivalent to retreiving the first 5 rows from the table 'tweets'
query = sql.select([tweets])
result = connection.execute(query).fetchall()
tweets_df = pd.DataFrame(result)
tweets_df.head(10)
```

```python
# Q.1. Retrieve the total number of tweets in the dataset.
tweets_count = sql.select([sql.func.count(col.text)])
total_tweets = connection.execute(tweets_count).scalar()
print(f"The total number of tweets in the dataset = {total_tweets}.")
```

```python
# Q.2. Find the number of distinct users (UserID) in the dataset.
distinct_users = sql.select([sql.func.count(col.UserID.distinct())])
distinct_count = connection.execute(distinct_users).scalar()
print(f"The number of distinct users in the dataset = {distinct_count}.")
```

```python
# Q.3. Calculate the average number of likes per tweet.
likes_average = sql.select([sql.func.round(sql.func.avg(col.Likes), 3)])
likes_avg = connection.execute(likes_average).scalar()
print(f"The average number of likes per tweet = {likes_avg}.")
```

```python
# Q.4. Identify tweets where the sentiment is 'Positive.' Display the TweetID and sentiment.
positive_tweets_query = sql.select(col.TweetID, sql.func.round(col.Sentiment, 2).label('Sentiment')).where(col.Sentiment > 0)
positive_tweets = connection.execute(positive_tweets_query).fetchall()
pd.DataFrame(positive_tweets).head(10)
```

```python
# Q.5. Count the number of tweets where IsReshare is true (1).
reshareTweetsQuery = sql.select([sql.func.count()]).where(col.IsReshare == 1)
reshareTweets = connection.execute(reshareTweetsQuery).scalar()
print(f"The number of tweets being reshared = {reshareTweets}.")
```

```python
# Q.6. List the top 5 users with the highest Reach. Display their UserID and Reach.
highestReachQuery = sql.select([col.UserID, sql.cast(col.Reach, sql.INT).label('Reach')]).order_by(col.Reach.desc()).limit(5)
highestReach = connection.execute(highestReachQuery).fetchall()
highestReach_df = pd.DataFrame(highestReach)
highestReach_df
```

```python
# Q.7. Find the most common language (Lang) used in tweets.
commonLanguageQuery = sql.select([col.Lang, sql.func.count().label('totalSpeakers')]).group_by(col.Lang).order_by(sql.func.count().desc()).limit(1)
commonLanguage = connection.execute(commonLanguageQuery).fetchone()
print(f"The most common language used in tweets = {commonLanguage['Lang']} ({commonLanguage['totalSpeakers']}).")
```

```python
# Q.8. Determine the average Klout score for male (Gender = 'Male') users.
maleKloutAvgQuery = sql.select([sql.func.round(sql.func.avg(col.Klout), 2)]).where(col.Gender == 'Male')
maleKloutAvg = connection.execute(maleKloutAvgQuery).scalar()
print(f"The average Klout score for male users = {maleKloutAvg}.")
```

```python
# Q.9. Retrieve tweets posted on weekdays (Monday to Friday).
weekDayTweetQuery = sql.select([tweets]).where(col.Weekday.in_(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']))
weekDayTweets = connection.execute(weekDayTweetQuery).fetchall()
pd.DataFrame(weekDayTweets).head(10)
```

```python
# Q.10. Identify tweets with a Klout score greater than 50. Display the TweetID and Klout.
greaterKloutQuery = sql.select([col.TweetID, sql.cast(col.Klout, sql.INT).label('Klout')]).where(col.Klout > 50)
greaterKlout = connection.execute(greaterKloutQuery).fetchall()
pd.DataFrame(greaterKlout).head(10)
```

```python
# Q.11. Count the number of tweets posted from the United States (Country = 'United States').
usTweetsQuery = sql.select([sql.func.count()]).where(col.Country == 'United States')
usTweets = connection.execute(usTweetsQuery).scalar()
print(f"The number of tweets posted from the United States = {usTweets}.")
```

```python
# Q.12. List tweets with the highest number of retweets. Display the TweetID and RetweetCount.
highestRetweetQuery = sql.select([col.TweetID, sql.cast(col.RetweetCount, sql.INT)]).order_by(col.RetweetCount.desc()).limit(1)
highestRetweet = connection.execute(highestRetweetQuery).fetchone()
print(f"The tweet ID {highestRetweet['TweetID']} has highest no. of retweets ({highestRetweet['RetweetCount']}).")
```

```python
# Q.13. Find tweets with sentiment 'Negative' and Klout score less than 40.

negativeKloutQuery = sql.select([col.TweetID, sql.cast(col.Klout, sql.INT).label('Klout'), sql.func.round(col.Sentiment, 2).label('Sentiment')]).where(col.Klout < 40).where(col.Sentiment < 0)
negativeKlout = connection.execute(negativeKloutQuery).fetchall()
pd.DataFrame(negativeKlout).head(10)
```

```python
# Q.14. Calculate the average Likes for tweets posted on weekends (Saturday and Sunday).
weekendLikesQuery = sql.select([sql.func.round(sql.func.avg(col.Likes), 3)]).where(col.Weekday.in_(['Saturday', 'Sunday']))
weekendLikes = connection.execute(weekendLikesQuery).scalar()
print(f"The average Likes for tweets posted on weekends = {weekendLikes}.")
```

```python
# Q.15. Retrieve tweets posted in the city of 'New York.'
newYorkQuery = sql.select([tweets]).where(col.State == 'New York')
newYorkTweets = connection.execute(newYorkQuery).fetchall()
pd.DataFrame(newYorkTweets).head(10)
```

```python
# Q.16. Identify tweets where Reach is greater than 1000. Display the TweetID and Reach.
highReachQuery = sql.select([col.TweetID, sql.cast(col.Reach, sql.INT).label('Reach')]).where(col.Reach > 1000)
highReach = connection.execute(highReachQuery).fetchall()
pd.DataFrame(highReach).head(10)
```

```python
# Q.17. Find the user (UserID) with the highest total engagement (sum of RetweetCount and Likes).
highEngagementQuery = sql.select([col.UserID, sql.cast((col.RetweetCount + col.Likes), sql.Integer).label('Engagement')]).where((col.RetweetCount + col.Likes) == sql.select([sql.func.max(col.RetweetCount + col.Likes)]).scalar_subquery())
highEngagement = connection.execute(highEngagementQuery).fetchone()
print(f"The UserID {highEngagement['UserID']} have the high engagement ({highEngagement['Engagement']}).")
```

```python
# Q.18. List tweets with sentiment 'Neutral' and Lang as 'English.'
neutralEnglishQuery = sql.select([col.TweetID, sql.cast(col.Sentiment, sql.INT).label('Sentiment'), col.Lang]).where(col.Lang == 'en').where(col.Sentiment == 0)
neutralEnglish = connection.execute(neutralEnglishQuery).fetchall()
pd.DataFrame(neutralEnglish).head(10)
```

```python
# Q.19. Calculate the total engagement (sum of RetweetCount and Likes) for each tweet.
totalEngagementQuery = sql.select([col.TweetID, sql.cast(col.RetweetCount + col.Likes, sql.INT).label('Engagement')]).group_by(col.TweetID)
totalEngagement = connection.execute(totalEngagementQuery).fetchall()
pd.DataFrame(totalEngagement).head(10)
```

```python
# Q.20. Retrieve tweets with sentiment 'Positive' or 'Neutral' and Lang as 'English' or 'Spanish.'

positiveNeutralEnglishQuery = sql.select([col.TweetID, sql.cast(col.Sentiment, sql.INT).label('Sentiment'), col.Lang]).where(col.Lang.in_(['en', 'es'])).where(col.Sentiment.in_(['Positive', 'Neutral']))
positiveNeutralEnglish = connection.execute(positiveNeutralEnglishQuery).fetchall()
pd.DataFrame(positiveNeutralEnglish).head(10)
```
